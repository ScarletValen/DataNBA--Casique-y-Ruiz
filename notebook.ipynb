```python
### librerias de interes a trabajar ###
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Cargar el dataset
df = pd.read_csv('30. NBA.csv')
```

```python


print("====================================================")
print("   RESULTADOS DEL ANÁLISIS DE DATOS - NBA GAMES     ")
print("====================================================\n")

# 1. Tasa de Victoria Local por Temporada
print("1. ¿Qué porcentaje de partidos gana el equipo local por temporada?")
consulta_1 = df.groupby('SEASON')['HOME_TEAM_WINS'].mean() * 100
print(consulta_1)
print("-" * 50)

```

```python
# 2. Diferencial de Puntos (Home - Away)
print("2. Diferencial de puntos promedio (¿Qué tanta ventaja saca el local?)")
df['PTS_DIFF'] = df['PTS_home'] - df['PTS_away']
consulta_2 = df.groupby('SEASON')['PTS_DIFF'].mean()
print(consulta_2)
print("-" * 50)

```

```python
# 3. Eficiencia en Triples (Home vs Away)
print("3. Comparativa de acierto en triples (%) entre Local y Visitante:")
consulta_3 = df.groupby('SEASON')[['FG3_PCT_home', 'FG3_PCT_away']].mean()
print(consulta_3)
print("-" * 50)

```

```python
# 4. Partidos de Alta Anotación (>110 puntos ambos)
print("4. Cantidad de partidos 'tiroteo' (Ambos equipos > 110 pts):")
alta_anotacion = df[(df['PTS_home'] > 110) & (df['PTS_away'] > 110)]
consulta_4 = alta_anotacion.groupby('SEASON').size()
print(consulta_4)
print("-" * 50)

```

```python
# 5. Correlación con la Victoria
print("5. ¿Qué estadística influye más para que el local gane?")
stats_cols = ['FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home', 'AST_home', 'REB_home', 'HOME_TEAM_WINS']
consulta_5 = df[stats_cols].corr()['HOME_TEAM_WINS'].sort_values(ascending=False)
print(consulta_5)
print("-" * 50)

```

```python
# 6. Top 10 Equipos más fuertes en Casa (por ID)
print("6. Top 10 IDs de equipos con mayor efectividad como local:")
consulta_6 = df.groupby('TEAM_ID_home')['HOME_TEAM_WINS'].mean().sort_values(ascending=False).head(10)
print(consulta_6)
print("-" * 50)

```

```python
# 7. El Efecto de Dominar los Rebotes
print("7. % de probabilidad de ganar si el local hace más rebotes:")
ganó_con_rebotes = len(df[(df['REB_home'] > df['REB_away']) & (df['HOME_TEAM_WINS'] == 1)])
total_con_mas_rebotes = len(df[df['REB_home'] > df['REB_away']])
consulta_7 = (ganó_con_rebotes / total_con_mas_rebotes) * 100
print(f"Probabilidad: {consulta_7:.2f}%")
print("-" * 50)
```

```python
# 8. Asistencias según el Resultado
print("8. Promedio de asistencias del local cuando Gana (1) vs cuando Pierde (0):")
consulta_8 = df.groupby('HOME_TEAM_WINS')['AST_home'].mean()
print(consulta_8)
print("-" * 50)

```

```python
# 9. Consistencia en Tiros Libres (Media y Desviación Estándar)
print("9. Estabilidad del tiro libre por temporada (Promedio y Variabilidad):")
consulta_9 = df.groupby('SEASON')[['FT_PCT_home', 'FT_PCT_away']].agg(['mean', 'std'])
print(consulta_9)
print("-" * 50)

```

```python
# 10. Temporadas más Ofensivas (Puntos Totales)
print("10. Ranking de temporadas por promedio de puntos totales por partido:")
df['TOTAL_PTS_GAME'] = df['PTS_home'] + df['PTS_away']
consulta_10 = df.groupby('SEASON')['TOTAL_PTS_GAME'].mean().sort_values(ascending=False)
print(consulta_10)
print("-" * 50)

```

```python
# 11. Partidos de Baja Anotación (La consulta que corregimos)
print("11. Partidos defensivos (Ambos equipos <= 95 puntos):")
baja_anotacion = df[(df['PTS_home'] <= 95) & (df['PTS_away'] <= 95)]
print(f"Total de partidos encontrados: {len(baja_anotacion)}")
print("Muestra de los últimos 5 partidos de este tipo:")
print(baja_anotacion[['GAME_DATE_EST', 'PTS_home', 'PTS_away']].tail())

print("\n====================================================")
print("            ANÁLISIS COMPLETADO                      ")
print("====================================================")
```
